- Create a dictionary to hold the article details. Initialize the dictionary values to an empty list for each key.  


- Loop through the articles
    - Assign the JSON values you will eventually store in a table to a variable
    - Convert the pub_date to a [datetime format](https://www.geeksforgeeks.org/python-pandas-to_datetime/)
    - Append the variable to the appropriate key for the dictionary initialized above
    - Print out the variable value
    - To delimit each article, print out a line of repeating non-alphanumeric characters of your choice

```python
article_data = {
    'main_headline': [],
    'web_url': [],
    'document_type': [],
    'pub_date': [],
    'word_count': [],
    'type_of_material': []
}

for article in docs:
    main_headline = article['headline']['main']
    print(main_headline)
    article_data['main_headline'].append(main_headline)
    web_url = article['web_url']
    print(web_url)
    article_data['web_url'].append(web_url)
    document_type = article['document_type']
    print(document_type)
    article_data['document_type'].append(document_type)
    pub_date = pd.to_datetime(article['pub_date'])
    print(pub_date)
    article_data['pub_date'].append(pub_date)
    word_count = article['word_count']
    print(word_count)
    article_data['word_count'].append(word_count)
    type_of_material = article['type_of_material']
    print(type_of_material)
    article_data['type_of_material'].append(type_of_material)
    print("*"*50)
```

### Make 10 requests to the API to collect 100 articles and save the results to the nyt_article table
A single API requests only returns 10 articles. Use the page parameter to "paginate" through the results.

The range function returns a range of numbers. range(10) returns 0-9.

Replace the placeholders denoted by \~ALL_CAPS\~.

```python
for article in range(10):
    print('Page:', article)
        
    api_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"
    params = {'api-key':"W5oqMomIPpg2cngEtkaYjNRM3MGtUCdj", 'page': article, 'begin_date':"20200101", 'end_date':"20201231", 'q':"Los Angeles"}

    # ~MAKE_API_REQUEST~
    api_request = requests.get(api_url, params=params)

    # ~DECODE_JSON_API_REQUESTS_RESULTS_AND_ASSIGN_TO_A_VARIABLE~
    api_response = json.loads(api_request.text)

    # ~ASSIGN_ARTICLES_LIST_TO_A_VARIABLE~
    docs = api_response['response']['docs']
    
    # ~INITIALIZE_DICTIONARY_TO_STORE_ARTICLE_DETAILS~
    article_data = {
        'main_headline': [],
        'web_url': [],
        'document_type': [],
        'pub_date': [],
        'word_count': [],
        'type_of_material': []
    }

    # ~LOOP_THROUGH_ARTICLES~
    for article in docs:
        
        # ~WEB_URL~
        web_url = article['web_url']
        article_data['web_url'].append(web_url)
        print(web_url)
        
        # ~MAIN_HEADLINE~
        main_headline = article['headline']['main']
        article_data['main_headline'].append(main_headline)
        print(main_headline)
        
        # ~DOCUMENT_TYPE~
        document_type = article['document_type']
        article_data['document_type'].append(document_type)
        print(document_type)
        
        # ~PUB_DATE~
        pub_date = pd.to_datetime(article['pub_date'])
        article_data['pub_date'].append(pub_date)
        print(pub_date)
        
        # ~WORD_COUNT~
        word_count = article['word_count']
        article_data['word_count'].append(word_count)
        print(word_count)
        
        # ~TYPE_OF_MATERIAL~
        type_of_material = article['type_of_material']
        article_data['type_of_material'].append(type_of_material)
        print(type_of_material)
        
        # ~PRINT_NONALPHANUMERIC_ARTICLE_DELIMITER~
        print("*" * 50)
        
    # ~ASSIGN_ARTICLES_DICTIONARY_TO_A_DATAFRAME~
    df = pd.DataFrame(article_data)
    
    # ~INSERT_DATAFRAME_INTO_nyt_article_TABLE_WITH_if_exists_append_OPTION~
    engine = create_engine('mysql+mysqldb://admin:sql_2021@assignment-02.cxmfqlltrzop.us-east-2.rds.amazonaws.com/assignment_02')
    df.to_sql('nyt_article', engine, index=False, if_exists='append')
    
```
